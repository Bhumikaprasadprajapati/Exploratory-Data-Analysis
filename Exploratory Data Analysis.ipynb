## 2. Histograms and boxplots for numeric Feature

```python

for column in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column],bins=10,color='lightgreen', edgecolor='black')
    plt.title(f'Histogram of {column}')  # Using f-string for formatting
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for column in df.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[column], color='blue')
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

## 3. Using pairplot for feature relationships

## 4. Pattern of Titanic Dataset

```python
for column in df.columns:
    plt.figure(figsize=(6, 4))
    sns.barplot(x = df[column],y = df["Survived"], color='lightgreen')
    plt.title(f'Barplot of {column}')
    plt.xlabel(column)
    plt.show()
```
